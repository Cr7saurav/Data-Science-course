```python
# Factorial
i=1
fact=1
while i<=10:
    fact*=i
    print(fact)
    i=i+1
    
```

```python
n=int(input("enter a number: "))
i=1
total=0
while i<=n:
    total+=i
    i+=1
    print("sum of the numbers from 1 to",n,"is",total)
```

```python
print("Welcome to My Store")
print("Type exit to finish your order")
count = 0
item = input("Enter Item: ")
while item.lower() != 'exit':
    count+=1
    print("Item" , count ,"added :", item)
    item = input("Enter Item: ")
```

# For Loop

```python
# single parameter for range (n) 
for i in range (5):
    print(i)
```

```python
# two parameters for range (start,stop)
for i in range (1,5):
    print(i)
```

```python
# multiple paramenters for range (start,stop,step)
for i in range (2,10,2):
    print(i)
```

```python
for i in range (5,0,-1):
    print(i)
```

```python
for i in range(1,6):
    print(" "*(5-i)+"*"*(2*i-1))
```

```python
sumOdd=0
sumEven=0
for i in range(1,11):
    if(i%2==0):
        sumEven=sumEven+i
    else:
        sumOdd=sumOdd+i
print("sumEven : ",sumEven)
print("sumOdd : ",sumOdd)
diff=sumEven-sumOdd
print("Difference of sumeven and sumodd from 1 to 10  : ",diff)
```
