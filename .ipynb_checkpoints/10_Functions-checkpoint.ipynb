# Recursive Function

```python
def factorial(n):
    if(n==1):
        return 1
    else:
        return (n* factorial(n-1))
```

```python
factorial(9)
```

# To add all the numbers between 1 to n

```python
def add(n):
    temp=0
    for i in range(1,n+1):
        temp+=i
    return temp
```

```python
num=int(input("Enter the Range For Sum From 1 to n:"))
sumu=add(num)
print(f"The Sum od range between 1 to {num} is :{sumu}")
```

# Map Function

```python
num=[1,2,3,4]
def double(x):
    return x*2
result= map(double,num)
print(list(result))
```

```python
num1=[1,2,3,4]
double1 = lambda x:x**2
result=map(double1,num1)
print(list(result))
```

```python
words=["coke","shoke","roke"]
upper=map(lambda u:u.upper(),words)
print(list(upper))
```

```python
item_prices=[100,250,80,120]
gst=map(lambda g:round(g*1.18,2),item_prices)
print(list(gst))
```


# Case Study: Bank Customer Transaction Management using function 

## Objective:

- Use a dictionary to **store and retrieve customer details** in a banking system.
- Implement functions to **add, update, and display** customer information.


## Industry Relevance:

In banking, customer details such as account number, name, and balance are stored in structured databases.

Using Python dictionaries, we can efficiently manage such records, simulating a small-scale banking application.



```python
customer={}
```

```python
def add_customer(account_no,name,amount):
    if account_no in customer:
        print("Account Already Exists")
    else:
        customer[account_no]={"Name": name , "Balance":amount}
        print("Customer Added Successfuly")
```

```python
def display_customer(account_no):
    if account_no in customer:
        print(f"Account Number:{account_no}")
        print(f"Name: {customer[account_no]['Name']}")
        print(f"Balance: {customer[account_no]['Balance']}")
    else:
        print("Account Not Exist")
```

```python
def update_balance(account_no,amount,transaction_type):
    if account_no in customer:
        if transaction_type == "Deposit":
            customer[account_no]["Balance"]+=amount
        elif transaction_type == "Withdraw" and customer[account_no]["Balance"]>=amount:
            customer[account_no]["Balance"]-=amount 
        else:
            print("Insufficient Balance")
        print("Transaction Successful")
    else:
        print("Account Not Exist")
    
```

```python
add_customer(101,'Saurav',5000)
```

```python
display_customer(101)
```

```python
update_balance(101,10000,"Deposit")
```

```python
display_customer(101)
```
