# Nested loop

```python
# Multiplication matrix
for i in range(1,4):
    for j in range(1,4):
        print(i*j,end="\t")
    print()
```

```python
# Right angled triange using star

for i in range (1,6):
    print("*"*i)
```

```python
## # 3x3 matrix sum
total=0
for i in range (1,4):
    for j in range(1,4):
        num=int(input(f"Enter element at position[{i},{j}]"))
        total+=num
        
print("The sum off all matrix elements is :" , total)
```

```python
## # 3x3 matrix sum
matrix=[]
total=0
for i in range (3):
    row=[]
    for j in range(3):
        num=int(input(f"Enter element at position[{i},{j}]"))
        row.append(num)
        total+=num
    matrix.append(row)
       
print("The sum off all matrix elements is :" , total)
for row in matrix:
    for element in row:
        print(element,end="\t")
    print()
        

```

```python
# break
for i in range(1,100):
    if i%7==0:
        print("First Multiple of 7 is:",i)
        break
```

```python
while True:
    num=int(input("Enter a Number:"))
    if num<0:
        print("You Entered a Negative Number")
        break
    print(num)
```

```python
# Continue
for i in range(1,10):
    if(i==5):
        continue
    print(i)
```

```python
for i in range (1,10):
    if i%2==0:
        continue
    print("Odd Numbers from 1 to 10:",i)
```
