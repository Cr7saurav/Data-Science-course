```python
# normal for
sq=[]
for i in range(1,101):
    sq.append(i**2)
print (sq)
```

```python
# list compre.
squa=[x**2 for x in range (1,101)]
print(squa)
```

```python
# list compre.
squa=[x**2 for x in range (1,101) if(x%2==0)]
print(squa)
```

```python
movies = [
    "Inception",
    "Interstellar",
    "The Dark Knight",
    "Avengers: Endgame",
    "Parasite",
    "Forrest Gump",
    "The Matrix",
    "Joker",
    "Titanic",
    "Gladiator",
    "The Shawshank Redemption",
    "Fight Club",
    "Pulp Fiction",
    "Avatar",
    "The Godfather"
]
```

```python
t=[]
for str in movies:
    if str.startswith('T'):
      t.append(str)
print(t)
```

```python
e=[]
for str in movies:
    if str.endswith('e'):
        e.append(str)
print(e)
```

```python
t=[str for str in movies if str.startswith('P')]
print(t)
```

```python
print(t)
```
