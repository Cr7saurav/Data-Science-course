# Comparison Operator

```python
# Equalto
age=18
print('age == 18:',age==18)
```

```python
# notEqualto
print("age!=18",age != 18)
```

```python
# grater than
print("age grater than 18" , age>18)
```

```python
# case study
marks= 65
if marks >=75 :
    category = "Distinction"
elif marks >=60 :
    category = "First Class"
elif marks >=40 :
    category = "Pass Class"
else :
    category = "Fail"
print(marks)
print(category)
```
