# Indexing and Slicing

## Indexing 
#### Positive(0 to len(str)-1)
#### Negative (-1 to -len(str))

```python
my_string="Saurav Yadav"
```

```python
print(my_string[0])
print(my_string[8])
```

```python
print(my_string[-1])
print(my_string[-5])
```

# Slicing
###### string[start:stop:step]

```python
s="saurav yadav"
print(s[0:1])
print(s[2:6])
print(s[:9])
print(s[2:11:2])
print(s[::-1])
```
