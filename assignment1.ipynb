# Employee Management System (EMS)
Objective - Create a simplified Employee Management System (EMS).
This project will cover control structures, functions, and object-oriented programming concepts to manage employee data.

```python
# Step 1
emp={
       101: {'name': 'Satya', 'age': 27, 'department': 'HR', 'salary': 50000},
       102: {'name': 'Rahul', 'age': 30, 'department': 'IT', 'salary': 60000}
}
```

```python
def add_emp():
    emp_id = int(input("Enter Emp_id"))
    if emp_id in emp:
        print("Employee Already Exists")
    else:
        name=(input("Enter Name:"))
        age=(int(input("Enter Age:")))
        department=(input("Enter Department:"))
        salary=(int(input("Enter Salary:")))
        emp[emp_id]= {'name': name,
                'age':age,
                'department':department,
                'salary':salary}
        print("Empl0yee Added Sucessfully")
        
```

```python
def view_employee():
    if not emp:
        print("No Employee Available")
    else:
        print("\nID\tName\tAge\tDepartment\tSalary")
        print("-" * 50)

        for emp_id, details in emp.items():
              print(f"{emp_id}\t{details['name']}\t{details['age']}\t"
              f"{details['department']}\t\t{details['salary']}")
```

```python
def search_employee():
    emp_id = int(input("Enter Employee ID to search: "))

    if emp_id in emp:
        emp1 = emp[emp_id]
        print("\nEmployee Found:")
        print(f"Name       : {emp1['name']}")
        print(f"Age        : {emp1['age']}")
        print(f"Department : {emp1['department']}")
        print(f"Salary     : {emp1['salary']}")
    else:
        print("Employee not found.")
```

```python
def main_menu():
    while True:
        print("\n===== Employee Management System =====")
        print("1. Add Employee")
        print("2. View All Employees")
        print("3. Search Employee")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add_emp()
        elif choice == '2':
            view_employee()
        elif choice == '3':
            search_employee()
        elif choice == '4':
            print("Thank you for using EMS. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

main_menu()
```
