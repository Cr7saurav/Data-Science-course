# CaseStudy

```python
# Dictionary for storing items and Stock
inventory={
    "rice":30,
    "suger":20,
    "wheat":40
}
```

```python
# Displaying all Itw=ems and thier quantity
for items,quantity in inventory.items():
    print("Item:",items.capitalize()," ","Stock:",quantity,"KG")
```

```python
# Adding new Item And Updating Exiting One
inventory["salt"]=10
print("Item Added Successfully")
inventory["suger"]+=10
print("Stock Upadted Successfully")
```

```python
for items,quantity in inventory.items():
    print("Item:",items.capitalize()," ","Stock:",quantity,"KG")
```

```python
# Item Availablity
item="salt"
if item in inventory:
    print(f"{item.capitalize()} in stock: {inventory[item]} KG")
else:
    print(f"{item.capitalize()} Not availabe")
```

```python
# Subset a<=b or a.issubset(b)
a<=b
a.issubset(b)
```

```python
A = {"Amit", "Saurav", "Rahul", "Neha", "Pooja"}
B = {"Saurav", "Ankit", "Neha", "Riya"}
C = {"Rahul", "Neha", "Kunal"}
```

```python
All_students = {"Amit", "Saurav", "Rahul", "Neha", "Pooja", "Ankit", "Riya", "Kunal", "Vikas"}
```

```python
enrolled=A|B|C
print("List of Student Enrolled in at least one Course :",enrolled)
```

```python
enrolled = A&B
print("Students enrolled in both Python and Data Science are:",enrolled)
```

```python
enrolled =A-B-C
print("Students enrolled only in Python are:",enrolled)
```
